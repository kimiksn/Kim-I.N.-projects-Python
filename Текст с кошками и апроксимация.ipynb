```python
def f(x):
    return np.sin(x/5)*np.exp(x/10)+5*np.exp(-x/2)
```

```python
def polynom1(x):
    return [1,x]
```

```python
a=np.array([polynom1(1),polynom1(15)])
b=[3.252216865271419,0.6352214195786656]
```
